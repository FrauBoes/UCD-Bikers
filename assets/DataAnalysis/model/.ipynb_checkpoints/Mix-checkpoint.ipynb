```python
from datetime import datetime
days=[]
hours=[]
weekday=[]
for index, row in df.iterrows():
    myNumber =int(row["TIME"]/1000)
    dt=datetime.fromtimestamp(myNumber)
    days.append(dt.month *100 + dt.day)
    hours.append(dt.hour)
    weekday.append(dt.weekday())
```

```python
df["days"]=days
df["hours"]=hours
df['weekday']=weekday
```

```python
df.dtypes
```

Convent the Station_number as categorical number.

```python
df['STATION_NUMBER'] = df['STATION_NUMBER'].astype('category')
df = df.drop(["TIME", "NUMBER_OF_SPACES"],axis = 1)
```

```python
import statsmodels.formula.api as sm
for i in range(7):
    for j in range(24):
        model_name =str(j+i *24)+''
        df_tem = df[(df["weekday"] == i)& (df["hours"] == j)]
        lm = sm.ols(formula = "NUMBER_OF_BIKES ~ weather_rain + weather_temperature + C(STATION_NUMBER)",data=df_tem).fit()
        lm.save(model_name)
```
